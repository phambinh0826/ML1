# Lecture 2 - Degree Centrality

## Degree Centrality

Đây có lẽ là chỉ số mạng xã hội đơn giản nhất để tính toán, và nó được sử dụng phổ biến trong nhiều ứng dụng.

## **Mục tiêu**
- Phân biệt giữa bốn chỉ số trung tâm cơ bản mà chúng ta sẽ nói đến
- Cách tính toán độ trung tâm theo bậc bằng tay như thế nào?


## **Tính trung tâm là gì và tại sao chúng ta nghiên cứu nó?**
Khi tôi trình bày mạng này, bạn nghĩ ai là người quan trọng nhất?

![Hình 5](fig5.png)

Tôi muốn bạn tạm dừng, nhìn vào nó một lúc và ghi lại cho mình những nút quan trọng nhất mà bạn nghĩ sẽ có tầm ảnh hưởng nhất.

Bạn có thể nhận ra mạng này là mạng lõi-ngoại vi (core-periphery network). Nó có một số tác nhân trung tâm, tôi tin rằng các tác nhân 2, 8, 5, 17, 18 và 39 là lõi, và sau đó tất cả các tác nhân khác đều kết nối với một trong các nhóm lõi đó.

Nhưng khi chúng ta nhìn vào ai là người quan trọng nhất, thì có một chút khó khăn. Nếu tôi nhìn vào một mạng có hình dạng không quá rõ ràng, có vẻ khó để đưa ra quyết định. Bạn có thể tạm dừng và thử lại bài tập này.

Bạn nghĩ ai là người quan trọng nhất? Giả sử bạn đã có cơ hội làm điều đó, điều gì sẽ xảy ra nếu không có câu trả lời đúng trong trường hợp này? Thực tế là rất khó để đưa ra câu trả lời trong bối cảnh sau.

![Hình 6](fig6.png)

Những gì chúng ta sẽ làm với các chỉ số trung tâm là tìm ra những cách khác nhau để đo lường ai là người trung tâm hay ai là người quan trọng trong mạng lưới. Điều này tương tự như khái niệm trung bình, trung vị, mode với dữ liệu truyền thống.

**Các phép đo trung tâm (Measures of Center)**

- **Trung bình (Mean - Average)**: 
  $$ \frac{\sum x_i}{n} $$ 
  (Tổng của tất cả các giá trị $x_i$ chia cho số lượng $n$)

- **Trung vị (Median)**: 
  Sắp xếp danh sách và sau đó xác định giá trị nằm ở giữa danh sách.

- **Mode (Mode)**: 
  Giá trị dữ liệu có tần suất cao nhất.

Tôi sẽ chỉ cho bạn sự tương đồng này với mạng lưới ngay sau đây.

Hãy dành một chút thời gian và làm bài trên nhé.

Đây là mô tả sơ lược, nhưng nó cho bạn hướng đi về cách bạn có thể tạo ra một trung bình là 6, trung vị là 5 và mode là 4. Tuy nhiên, phần quan trọng là tất cả các phép đo này, mặc dù tính toán khác nhau và khác nhau trên một tập hợp 10 số, đều là các phép đo trung tâm. 

Chúng là các phép đo cố gắng xác định trung tâm. Chúng ta biết rằng khi chúng ta có một phân phối đối xứng, có một đường cong chuông, có số lượng lớn bằng nhau giữa các giá trị cao và thấp, thì trung bình hoạt động rất tốt và chúng ta thường xuyên sử dụng trung bình.

Tại sao chúng ta sử dụng trung bình nhiều vậy? Chúng ta sử dụng trung bình vì có lẽ nó dễ tính toán hơn trung vị, nhiều khi vì bạn không phải sắp xếp dữ liệu. Nhưng vấn đề khác với trung bình là có một định lý gọi là **định lý giới hạn trung tâm** trong thống kê, đó là nếu tôi lấy trung bình của 30 số trở lên, tôi sẽ biết phân phối của trung bình đó. Ngay cả khi các số có phân phối rất lệch, nếu tôi đang lấy mẫu và lấy trung bình của chúng, tôi biết phân phối, điều này cho phép tôi có suy luận thống kê về việc so sánh hai nhóm hoặc sự thay đổi. Chúng ta thường sử dụng trung bình vì chúng ta có thể sử dụng các tính chất khác mà chúng ta không dành thời gian để dạy mọi người.

Nhưng khi bạn có dữ liệu bị lệch rất nhiều, điều này được sử dụng nhiều trong kinh tế học hoặc chính trị vì có rất nhiều người giàu và nghèo, và khi bạn có một quần thể lệch như vậy, nếu chúng ta lấy giá trị trung bình, chúng ta sẽ suy luận hoặc ước đoán về một cái gì đó không thực sự đại diện cho bất kỳ nhóm nào.Chúng ta sử dụng trung vị như một cách hiệu quả hơn để đo lường trung tâm khi có cấu trúc cao hoặc sự lệch trong dữ liệu.

Mode không được sử dụng nhiều, mặc dù có thể là cách dễ nhất để đo lường. Tương tự, chúng ta có thể nhìn vào mạng lưới kiểu mẫu này từ Dave Crackhart.

Bây giờ, mạng lưới, chỉ để cung cấp cho bạn một số thông tin nền, là một trong những sinh viên của Dave Crackhart, Red, và ở dưới đây.

Red không phải là tôi. Tôi biết tôi có tóc đỏ, nhưng không phải là tôi. Anh ta là một sinh viên quân sự trong lớp của Dave Crackhart, và bài tập của anh ấy sau khi tốt nghiệp là tại Lầu Năm Góc. Có một mạng lưới này, và Sam, ở góc trên bên phải gần với 'Degree', là tướng một sao người đã phụ trách nhóm này tại Lầu Năm Góc và hai nhóm khác nhau đã được hợp nhất lại.

Sam đã phụ trách, vì vậy Red nghĩ anh ấy sẽ chỉ cho nhóm phân tích mạng xã hội. Cuối cùng điều đó không phải là một ý tưởng hay. Nhưng chúng ta có được những mạng lưới kiểu mẫu tuyệt vời từ bài tập này với Dave.

Những gì chúng ta thấy ở đây là một mạng lưới có cấu trúc, vì vậy có một nhóm ở góc trên bên phải và một nhóm ở góc dưới bên trái, và bạn thấy *"Conrad"* ở giữa. Khi chúng ta nhìn vào các thuật ngữ **degree, betweeness, closeness, eigenvector,** chúng là những cách khác nhau để đo lường trung tâm.

> 1. Khi tôi nhìn vào **degree**, tôi thấy Linda có nhiều kết nối vật lý nhất trong mạng lưới. Cô ấy kết nối với nhiều người hơn bất kỳ ai khác. Tôi nghĩ Roger đứng thứ hai, nhưng Linda có nhiều kết nối hơn bất kỳ ai khác. Nhưng Linda là trung tâm trong nhóm ở góc trên bên phải và cô ấy không gần với những người trong nhóm dưới bên trái.
>
> 2. Nếu bạn nhìn vào Roger, Roger gần hơn với tất cả mọi người khác trung bình. Nếu chúng ta đo chiều dài của các đường đi ngắn nhất giữa tất cả các cặp nút và sau đó tính trung bình, chúng ta sẽ thấy rằng Roger có khoảng cách ngắn nhất trung bình đến tất cả mọi người khác trong mạng lưới. Đó là một loại trung tâm khác - **closeness**.
>
> 3. Sau đó, người nổi bật với mọi người là Conrad vì Conrad là một cầu nối đơn lẻ giữa hai nhóm, và anh ấy có mức độ **betweeness** cao. Nếu bạn nhìn vào đường đi ngắn nhất và nói, tần suất các nút rơi vào đường dẫn đó bao nhiêu lần, Conrad nằm ở giữa nhiều đường đi ngắn nhất giữa các tác nhân hơn bất kỳ ai khác.
>
> - Đây là những cách khác nhau để nhìn vào trung tâm.
>
> 4. **Eigenvector centrality** được sử dụng trong thuật toán pagerank của Google. Điều nó làm là nói rằng nếu bạn đi bộ ngẫu nhiên qua mạng lưới, không có ràng buộc, bạn chỉ đi bộ ngẫu nhiên, xác suất mà một nút nhất định sẽ xuất hiện trên đoạn đường đi ngẫu nhiên đó là bao nhiêu?

Điều xảy ra khi bạn có cấu trúc là khả năng bạn sẽ đi từ một nhóm này sang nhóm khác là khá thấp so với khả năng bạn sẽ ở trong một nhóm duy nhất. Trong một mạng lưới như thế này, vì nhóm ở góc trên bên phải có một chút lớn hơn, nên khả năng một nút chỉ lưu thông trong nhóm đó ở góc trên bên phải là cao hơn, và đó là lý do bạn sẽ thấy một tác nhân trung tâm đặc biệt trong nhóm đó sẽ nổi bật. Chúng ta thấy rằng các **eigenvectors** cao nhất có xu hướng có sự tương quan cao với **degree** cao nhất trong nhóm đó ở góc trên. Nhưng nó được sử dụng để có một cảm giác nếu chúng ta thực hiện các chuyển động ngẫu nhiên qua mạng lưới, một người sẽ có tần suất như thế nào hoặc mức độ trung tâm hay ảnh hưởng ra sao?

Trên đây là bốn phép đo trung tâm cơ bản, và chúng ta sẽ học cách tính toán **degree centrality** trong bài học này.


## **Network Measures of Centrality**

>**Degree** là số lượng cạnh kết nối với một nút.

>**Betweenness** là mức độ mà nó nằm trên đường đi ngắn nhất giữa các nút.

>**Closeness** là trung bình khoảng cách ngắn nhất đến tất cả các nút khác trong mạng lưới, và

>**Eigenvector** đo lường mức độ mà một nút kết nối với những nút có ảnh hưởng khác.

Không phải là ai đang kết nối với bạn mà là bạn đang kết nối với ai hoặc bạn có kết nối với những người có ảnh hưởng không? Những phép đo này cho tôi biết gì?

## **What do the Measures Tell Me?**

Nó hơi khác một chút. **Degree** là mức độ tiếp xúc với mạng lưới. 
- Khi bạn nghĩ đến việc phát tán một tin tức gây sốc hay một cái gì đó như vậy, nút có nhiều kết nối sẽ nhận thông tin ban đầu và phát tán ra mạng lưới đến nhiều người nhất.
- Tuy nhiên, chúng ta thấy rằng những người có **degree** cao không có nhiều sức ảnh hưởng. Họ không thay đổi được suy nghĩ, không nắm quyền lực thực sự để động viên hoặc thuyết phục, nhưng đó là cách để đưa thông tin ra ngoài. Ngay cả khi khả năng thuyết phục của bạn thấp, bạn bắt đầu với nhiều người hơn.
- Đó là điều **degree** cho bạn biết là tiếp xúc với mạng lưới.

**Betweenness**, đây là ý tưởng về quyền lực không chính thức.
- Đây là ý tưởng về việc trung gian hóa kiến thức và nguồn lực trong toàn tổ chức.
- Các bạn đã từng ở trong những tổ chức mà có có người không chịu trách nhiệm chính, nhưng họ biết chỗ nào có mẫu đơn, họ biết cách giải quyết công việc. Họ tham gia vào công việc và biết cách giải quyết vấn đề. Những người đó xây dựng quyền lực không chính thức, và họ có rất nhiều kết nối ở những nơi khác.
- Sự trung gian không chính thức, việc canh giữ giữa một trung tâm như Conrad có khả năng trung gian hóa kiến thức từ nhóm ở phía bên phải sang nhóm ở phía bên trái dưới cùng. Đôi khi, đó có thể là một vị trí quyền lực lớn. Những lúc khác nó có thể khiến bạn kiệt sức về mặt nhận thức, khi bạn là người đi tới giữa hai nhóm.
- Trong trường hợp của Conrad, anh ấy cuối cùng đã nghỉ việc, và tổ chức tan rã vì điều đó, và tướng quân về hưu, đó không phải là một câu chuyện thành công. Nhưng đó là **betweenness**.

**Closeness** thường được sử dụng để truyền bá thông tin vì nó giúp thông tin được tiếp cận nhanh chóng.
- Giống như khi tin đồn đang lan truyền qua mạng lưới hoặc bạn muốn bắt đầu một tin đồn lan truyền qua mạng lưới.
- Nếu bạn có thể tìm ra điểm hoặc người có **closeness** cao, đó là điểm lan truyền thông tin nhanh chóng.

Cuối cùng, như đã đề cập về **eigenvector**, nó giúp bạn kết nối với những nút có degree cao.
- Đôi khi, khi bạn có thể tìm ra một nút có **eigenvector centrality** cao, nhưng degree thấp, bạn đang tìm thấy một nút tuyệt vời để phát tán tin đồn hoặc thông tin sai lệch mà không nhất thiết sẽ bị phát hiện, vì hầu hết các thuật toán mà mọi người sử dụng đều tìm các nút có degree cao vì đó là phép đo dễ tính toán nhất.

Nếu bạn có thể tìm thấy một nút có degree thấp nhưng có một trong các giá trị khác cao, đó sẽ là một mục tiêu thú vị và nếu bạn muốn giỏi hơn trong việc phát hiện thông tin sai lệch, bạn cần phải tìm những cách thông minh hơn để phát hiện những nút có thể bắt đầu lan truyền tin đồn mà không nhất thiết bị phát hiện.

## **Degree centrality**
Degree đơn giản là số lượng các cạnh kết nối với một nút.

![Hình 8](fig8.png)

Nếu tôi nhìn vào mạng lưới này ở đây như một ví dụ,

Bước đầu tiên tôi sẽ làm là ghi lại phương trình ở đây,
$$ C_{D_i} = \frac{\sum_{j=1}^{n} a_{ij}}{n - 1} $$

Đây là **degree centrality** cho một nút nhất định
- I cho trước sẽ là tổng các phần tử của hàng trong ma trận kề, và chúng ta sẽ chia cho $n-1$. Điều này trông như thế nào?

Hãy để tôi tạo một ma trận kề. 

![Hình 9](fig9.png)

Một điều nữa về ma trận kề là đường chéo của ma trận kề là không xác định.
> Điều này có nghĩa là gì khi một tác nhân kết nối với chính nó? Điều này có nghĩa là gì khi một tác nhân là bạn của chính nó? Trong một lập luận, bạn có thể nói, tôi thực sự biết tôi là ai. Một lập luận khác là, tôi không phải như vậy, tôi có đang dành thời gian vô hạn với chính mình không? Đó là một khái niệm kỳ lạ.

Trong phân tích mạng xã hội, chúng ta chỉ coi đường chéo chủ yếu là không xác định.
> Nếu tôi nhìn vào một tác nhân, tôi có thể thấy ở đây Tác nhân 1 chỉ kết nối với Tác nhân 4, không phải với các tác nhân khác. Tôi có thể làm điều tương tự cho Tác nhân 2. Điều tương tự với Tác nhân 3, Tác nhân 4 kết nối với ba tác nhân khác nhau, và sau đó là Tác nhân 5. Đó là ma trận kề.

Từ ma trận kề, tôi có thể đơn giản chỉ cần cộng lại các giá trị trong hàng và tôi có thể có giá trị ở đó, sau đó tôi sẽ chia cho $n-1$.
> Có bao nhiêu liên kết khả thi nếu một nút kết nối với tất cả mọi người? Có năm tác nhân trong mạng lưới này.

Nếu một nút kết nối với tất cả các tác nhân còn lại, số lượng tối đa mà nó có thể kết nối là bốn nút khác, vì nó sẽ không kết nối với chính nó. Do đó, $n-1$ là bốn, và điều đó cho tôi một **degree** là 1/4, 1/4, 2/4, 3/4 và 1/4. Điều đó trở thành **degree** của mỗi nút trong mạng lưới.
